```python
import json
import time

#reading/loading records from json to string
fd = open("records.json", "r")
k = fd.read()
fd.close()

records = json.loads(k)

# Menu section
print("-"*15 + "MENU" + "-"*15)
for i in records:
    print(i, ":", records[i]['Name'], "|" ,str(records[i]['Price'])+chr(8377), "|", records[i]['Qn'])
print("-"*34,"\n\n")

#user_input
ui_Name     = input("Enter User Name                    : ")
ui_Phn      = int(input("Enter Phone Number                 : "))
ui_Email    = input("Enter Mail ID                      : ")

cart = []

while True:
    ui_Product  = input("Enter Product ID                   : ")
    if ui_Product in records.keys():
        if records[ui_Product]['Qn'] == 0:
            print("""Sorry!!! We don't have stock""")
        else:
            ui_Quantity = int(input("Enter Product Quantity             : "))
            if records[ui_Product]['Qn'] >= ui_Quantity:

                #sale details
                sale = ui_Name + ',' + str(ui_Phn) + ',' + ui_Email + ',' + ui_Product + ',' + str(ui_Quantity) + ',' + str(records[ui_Product]['Price']) + ',' + str(ui_Quantity * records[ui_Product]['Price']) + ',' + time.ctime() + "\n"

                l = [ui_Product, ui_Quantity, records[ui_Product]['Price'], ui_Quantity * records[ui_Product]['Price']]

                cart.append(l)

                #updating the product Quantity
                records[ui_Product]['Qn'] -= ui_Quantity

            else:
                print("Sorry! We don't have enough stock")
                print("We have", records[ui_Product]['Qn'], "Quantity")
                print("Would you like to purchase?")
                ch = input("Press Y/N                          : ")
                if ch == 'Y' or ch == 'y':

                    #sale details
                    sale = ui_Name + ',' + str(ui_Phn) + ',' + ui_Email + ',' + ui_Product + ',' + str(records[ui_Product]['Qn']) + ',' + str(records[ui_Product]['Price']) + ',' + str(records[ui_Product]['Price']*records[ui_Product]['Qn']) + ',' + time.ctime() + "\n"

                    l = [ui_Product, records[ui_Product]['Qn'], records[ui_Product]['Price'], ui_Quantity * records[ui_Product]['Price']]

                    cart.append(l)

                    #updating the product Quantity
                    records[ui_Product]['Qn'] = 0
        ch = input("If you want to purchase more (Y/N)?: ")
        if ch == 'Y' or ch == 'y':
            continue
        else:
            break
                
        #updating the sales details
        fd = open("Sales.txt", "a")
        fd.write(sale)
        fd.close()
            
    else:
        print("We don't have that product!!! Purchase another one...")
        
    

# Billing section
print("\n\n")
print("-" * 31 + "BILL" + "-" * 31)
print(f"{'product_name':<15} | {'product_quantity':<17} | {'product_price(₹)':<17} | {'Total(₹)':<5}")
print("-" * 66)
Grand_Total = 0
for i in cart:
    product_name = records[i[0]]['Name']
    product_quantity = i[1]
    product_price = i[2]
    Total = i[3]
    Grand_Total += i[2]
    
    print(f"{product_name:<15} | {product_quantity:<17} | {product_price:<17} | {Total:<5}")
print("-" * 66)
print(f"{'Grand_Total':<57} {Grand_Total:<2}{chr(8377)}")
print("-" * 66)
        
#converting dict into str and update in records.json
js = json.dumps(records)
fd = open("records.json", "w")
fd.write(js)
fd.close()
```
